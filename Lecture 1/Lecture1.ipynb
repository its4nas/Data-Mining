```python
data_set = pd.read_csv('pokemon_data.csv')
data_set.head(6)

#Kaggle
```

```python
data_set.info()
```

```python
data_set.head()
```

```python
#drop null in source data
data_set.dropna(inplace=True)
data_set.head()
```

```python
#drop null in new variable
new_df = data_set.dropna()
new_df.head()
```

```python
#fill null
x = new_df['Speed'].mean()
x
new_df["Speed"].fillna(x,inplace=True)
new_df.head(6)
```

```python
#change value based on condition
for x in new_df.index:
    if new_df.loc[x]["HP"] > 70:
        new_df.loc[x, "HP"] = new_df["HP"].mode()[0]

new_df
```

```python
#delete duplicate data // the whole row
new_df.drop_duplicates()
new_df
```
