```python
import numpy as np
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score, calinski_harabasz_score
from mpl_toolkits import mplot3d
```

```python
df_cluster = pd.read_csv("FIFA18_modified_clustering.csv")
```

```python
df_cluster
```

```python
df_cluster.plot(kind='scatter',
               x='Overall',
               y='Value',
               figsize=(20,20))
```

```python
dbscan = DBSCAN(eps=0.033,                          #1/2
               min_samples=22,                  # numberv of points
               algorithm='brute',
               p=0.9)
preds = dbscan.fit_predict(df_cluster)
```

```python
silhouette_score(df_cluster, preds)       # make value between -1 and 1 for quality of data
```

```python
df_cluster.plot(kind='scatter', x='Overall', y='Value', c=preds, cmap='Accent_r', figsize=(20,20))
pass
```
