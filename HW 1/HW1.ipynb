```python
import pandas as pd
```

```python
data_set = pd.read_csv('FIFA18_official_data.csv')
data_set.head(10)
```

```python
pd.set_option('display.max_columns', None)
```

```python
no_empty = data_set.dropna()
no_img = data_set.drop(columns=['Photo','Flag','Club Logo','Position','Loaned From'])
no_img.head(21)
# no_empty.head()
```

```python
no_img.drop_duplicates()
no_img
```

```python
for x in no_img.index:
    if no_img.loc[x]["Finishing"] < 50:
        no_img.loc[x, "Finishing"] = no_img["Finishing"].mode()[0]

no_img.head(10)
```

```python
# no_img = no_img.fillna(1000)
df_without_missing = no_img.dropna()
```

```python
df_without_missing.to_csv('FIFA18_modified.csv',index=True)
```
